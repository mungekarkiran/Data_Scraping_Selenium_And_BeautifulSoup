```python

os.makedirs(IMAGE_DIR, exist_ok = True)

for ele in range(len(element)):
    
    element_text_list = element[ele].text.split('\n')
    element_tag_a_list = element[ele].find_elements(By.TAG_NAME, "a")
    element_tag_img_list = element[ele].find_elements(By.TAG_NAME, "img")

    print(len(element_text_list), len(element_tag_a_list), len(element_tag_img_list))
    
    if len(element_tag_a_list) != 0:
        
        for data in range(len(element_text_list)):
            
            Name_of_Disease = element_text_list[data]
            Disease_Page_URL = element_tag_a_list[data].get_attribute("href")
            Image_of_Diseases = element_tag_img_list[data].get_attribute("src")
            
            print(Name_of_Disease, end = " | ")
            print(Disease_Page_URL, end = " | ")
            print(Image_of_Diseases, "\n")
            
            image_name = re.sub('\W+','', Name_of_Disease)+".jpg"
            image_path = os.path.join(IMAGE_DIR,image_name)
            save_image(Image_of_Diseases, image_path)
            
        
    else:
        pass
    
    
```

```python
def mystery(n):
    v = 0
    for i in range(1, n+1):
        v1 = ''
        for j in range(i):
            v1 += str(i)
        v += int(v1)
    
    return v

print(mystery(1))
print(mystery(2))
print(mystery(3))
print(mystery(4))
print(mystery(5))
```
